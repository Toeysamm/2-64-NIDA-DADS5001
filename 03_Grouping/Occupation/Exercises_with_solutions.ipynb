### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

> เราต้องการที่จะ groupby occupation แล้วไปหา gender ที่เป็นเพศ 'M' (Male) แล้วไปหา ratio ของ 'M' ((M/total)x100)

```python
# create a function เพื่อต้องการให้ M เป็น 1 และ F เป็น 0
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

# apply the function to the gender column and create a new column (gender_n)
users['gender_n'] = users['gender'].apply(gender_to_numeric) #.apply ใช้ช่วยในการ เรียก function มาใช้


a = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100 
#ใช้ gender_n.sum() เพราะว่าเราสามารถหาได้จาก M เป็น 1 แล้ว (รู้ว่ามี Male กี่คน) เพื่อหาคำตอบได้
#users.occupation.value_counts() หมายความว่า นับจำนวน member ทั้งหมดใน columns occupation 

# sort to the most male 
a.sort_values(ascending = False)

#a.sort_values
```

### Step 8.  For each occupation present the percentage of women and men

8.1

```python
# create a data frame and apply count to gender ต้องหาจำนวนของแต่ละเพศในแต่ละอาชีพก่อน
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
gender_ocup
```

8.2

```python
# create a DataFrame and apply count for each occupation หาจำนวนเพศทั้งหมดของแต่ละอาชีพ
occup_count = users.groupby(['occupation']).agg({'gender':'count'})
occup_count
```

8.3

```python
# divide the gender_ocup per the occup_count and multiply per 100
#นำ 8.1 มาหาค่าร่วมกับ 8.2 เป็นคำตอบ
occup_gender = gender_ocup.div(occup_count) * 100 #.div คือหาร
occup_gender
```

8.3 (alternative)

```python
# OR

(gender_ocup/occup_count)*100
```
